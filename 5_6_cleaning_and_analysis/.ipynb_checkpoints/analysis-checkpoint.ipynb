Import pandas and set constant values

```python
import pandas as pd
import math as m
import matplotlib.pyplot as plt

viet_super = {'Winmart', 'Co.op'}
rus_super = {'Pyaterochka', 'Lenta'}

all_categories = {'bread', 'cucumber', 'tomato', 'cabbage', 'onion', 'carrot', 'eggplant', 'banana', 'orange', 'mango', 'milk', 'condensed_milk', 'tofu', 'egg', 'yogurt', 'pork_leg', 'chicken_fillet', 'spaghetti', 'buckwheat', 'rice', 'black_tea', 'sugar', 'green_tea', 'ground_coffee', 'salt', 'sunflower_oil', 'water', 'fish_sauce', 'rice_noodles', 'soybean_oil', 'water_spinach'}
per_lit_categories = {'fish_sauce', 'milk', 'water', 'sunflower_oil', 'soybean_oil'}
per_piece_categories = {'egg'}
per_kg_categories = all_categories - per_lit_categories - per_piece_categories
vegetables = {'cucumber', 'tomato', 'cabbage', 'onion', 'carrot', 'eggplant', 'water_spinach'}
fruit = {'banana', 'orange', 'mango'}
staples = {'spaghetti', 'buckwheat', 'rice', 'rice_noodles'}
oils = {'sunflower_oil', 'soybean_oil'}
dairy = {'milk', 'yogurt'}
meat = {'pork_leg', 'chicken_fillet'}
drinks = {'green_tea', 'black_tea', 'ground_coffee'}

exchange_rates = {
    'Vietnam': 0.000039214,
    'Russia': 0.011056 
}
```

Import data from CSV.
I also want to reorganize the table:
- add 'country' column
- change column order
- combine 'name' and 'uom' into a single string like "name (uom)"
- have new column 'quantity' that would specify unit of price normalization (kg, liter or unit). convert g and ml into kg and l, so it is not misleading
- combine weight, number_of_units and volume into a single column (corresponding prices into another single column)
- rename price columns into supermarket_price and normalized_price

```python
clean_products = pd.read_csv('clean_products-2025-03-12-local.csv')

# add country column
clean_products['country'] = clean_products.supermarket.apply(lambda x: 'Vietnam' if x in viet_super else 'Russia') 

# rename price and product_type columns
clean_products = clean_products.rename(columns={'price': 'supermarket_price', 'product_type': 'product_category'})

# combine name and uom
clean_products['product_name'] = clean_products.name + ' (' + clean_products.uom.fillna('') + ')'

# add unit type column
def assign_unit(product_category):
    if product_category in per_lit_categories:
        return 'l'
    elif product_category in per_piece_categories:
        return 'piece'
    else:
        return 'kg'

clean_products['unit'] = clean_products.product_category.apply(assign_unit)

# combine quantities into a single column, convert grams and milliliters
clean_products['quantity'] = clean_products[['weight', 'number_of_units', 'volume']].apply(lambda row: row.dropna().values[0], axis=1)
clean_products.loc[clean_products.unit.isin(['kg', 'l']), 'quantity'] /= 1000

# do the same for normalized prices
clean_products['normalized_price'] = clean_products[['price_kg', 'price_unit', 'price_lit']].apply(lambda row: row.dropna().values[0], axis=1)

# convert prices into USD, add supermarket_price_usd column
clean_products['supermarket_price_usd'] = clean_products.supermarket_price * clean_products.country.map(exchange_rates)
clean_products['normalized_price'] = clean_products.normalized_price * clean_products.country.map(exchange_rates)

# drop unnecessary columns and reorganize them
clean_products = clean_products.iloc[:, [12, 5, 4, 13, 2, 17, 15, 14, 16]]

# add one missing buckwheat entry
clean_products.loc[3000] = ['Vietnam', 'Winmart', 'buckwheat', 'Hạt kiều mạch hiệu Goodwill', 85600, 3.3567184, 0.8, 'kg', 4.195898]
# exclude a wrong chicken_fillet entry
clean_products = clean_products.drop(428)

# clean_products.to_csv('total-products-list.csv', index=False)
clean_products
```

### Outliers check, summary and basic plots

**Important notes:**
1. The distribution inside each category of products in each country is vastly different. Some resemble skewed normal distribution, some are completely irregular, some - have distinct clusters. Therefore, I will handle those categories differently
   1. **For categories that have continuous relatively dense distribution, both skewed or irregular** - I will drop significant outliers using IQR with log transformation method.
   2. **For categories with distinct clusters or just 1 or 2 data point** - I will keep all the data.
   3. Overall, I will decide on options 1 or 2 based on the price distribution histogram
2. I will use **a median** as the primary parameter for comparison as it works better then a mean with not-normal distributions

So, let's start with histograms. I will store all analysis tables and plots in a summary dictionary by category

```python
data_summary = {}    # create a dictionary

for category in all_categories:
    
    # extract data for selected category and drop irrelevant columns
    category_data = clean_products[clean_products.product_category == category]
    category_data = category_data.iloc[:, [0, 1, 2, 3, 8]]
    category_data_ru = category_data.loc[category_data.country == 'Russia']
    category_data_vn = category_data.loc[category_data.country == 'Vietnam']

    # add histograms
      # calculate number of bins as square root of number of data points
    bins_ru = 1 if len(category_data_ru) == 0 else m.ceil(m.sqrt(len(category_data_ru)))
    bins_vn = 1 if len(category_data_vn) == 0 else m.ceil(m.sqrt(len(category_data_vn)))
    
      # plot two histograms together
    fig, (ax1, ax2) = plt.subplots(2, 1)
    category_data_ru.normalized_price.hist(bins=bins_ru, ec='k', ax=ax1)
    ax1.set_title('Russia')
    category_data_vn.normalized_price.hist(bins=bins_vn, ec='k', ax=ax2)
    ax2.set_title('Vietnam')
    plt.tight_layout()
    plt.close('all')
    
    # save data and the plot into a dictionary
    data_summary[category] = {
        'hist': fig
    }
```

I have identified which categories to check for outliers and which not. Now we can create summary tables and plots for all of them. Histograms will be updated if the outliers are to be excluded.

```python
# clustered categories that will skip outliers check
skip_outliers_ru = ['banana', 'eggplant', 'tomato', 'ground_coffee', 'fish_sauce', 'condensed_milk', 'yogurt', 'carrot', 'cucumber', 'water_spinach', 'soybean_oil']
skip_outliers_vn = ['black_tea', 'eggplant', 'tomato', 'ground_coffee', 'chicken_fillet', 'condensed_milk', 'yogurt', 'carrot', 'cabbage', 'buckwheat', 'tofu']
```

```python
# outliers check function. Takes a Dataframe, adds two columns: log_price (log10 of normal. price) and is_outlier (True if outlier)
def outliers_check(data):

    # identify outliers with IQR method with log transformation
    # calculate quantiles, IQR and boundaries
    data['log_price'] = data.normalized_price.apply(m.log10)    # log10 for prices
    log_iqr = data.log_price.agg(          # Q1 and Q3
        q1=lambda x: x.quantile(0.25),
        q3=lambda x: x.quantile(0.75)
    )
    log_iqr['iqr'] = log_iqr.q3 - log_iqr.q1    # IQR
    log_iqr['lower_bound'] = log_iqr.q1 - log_iqr.iqr * 1.5
    log_iqr['upper_bound'] = log_iqr.q3 + log_iqr.iqr * 1.5     # boundaries
    
    # find outliers (points outside boundaries)
    data['is_outlier'] = data.apply(
        lambda row: True if (row.log_price < log_iqr['lower_bound']) or (row.log_price > log_iqr['upper_bound']) else False,
        axis=1
    )

    # drop unnecessary log_price column
    data.drop(['log_price'], axis=1, inplace=True)
```

```python
# summary stats function. Takes a dataframe, calculates count, mean and std, median, quantiles and IQR, min and max, outliers (if applicable)
def summary_stats(data):

    # exclude outliers for correct stats
    data_no_out = data.loc[~data['is_outlier']]
    
    # calculate all the functions and aggregate them into a DF right away 
    summary = data_no_out.groupby('country').normalized_price.agg(
        count=lambda x: None, # placeholder, will be calculated below
        outliers=lambda x: None, # placeholder
        outliers_percent=lambda x: None, # placeholder
        average='mean',
        std='std',
        p10=lambda x: x.quantile(0.1),
        q1=lambda x: x.quantile(0.25),
        median='median',
        q3=lambda x: x.quantile(0.75),
        iqr=lambda x: x.quantile(0.75) - x.quantile(0.25),
        min='min',
        max='max',
    ).reset_index()
    
    summary = summary.set_index('country')    # use countries as index
    # fill in the missing count and outliers
    summary['count'] = data.groupby('country').product_name.count()
    summary['outliers'] = data.groupby('country').is_outlier.sum()
    summary['outliers_percent'] = summary['outliers'] / summary['count'] * 100
    
    return summary
```

```python
# loop through the categories
for category in all_categories:

    # extract data for selected category and drop irrelevant columns (once again, I prefer repeating it in order to avoid data rewriting)
    category_data = clean_products[clean_products.product_category == category]
    category_data = category_data.iloc[:, [0, 1, 2, 3, 8]]
    category_data_ru = category_data.loc[category_data.country == 'Russia'].copy()
    category_data_vn = category_data.loc[category_data.country == 'Vietnam'].copy()

    # add 'is_outlier' column with False by default
    category_data_ru['is_outlier'] = False
    category_data_vn['is_outlier'] = False

    # if applicable and the list of products isn't empty, check for outliers
    if (category not in skip_outliers_ru) and (len(category_data_ru) > 0):

        # check for outliers
        outliers_check(category_data_ru)
        
    # same check for VN
    if (category not in skip_outliers_vn)  and (len(category_data_vn) > 0):

        outliers_check(category_data_vn)
        
    category_data = pd.concat([category_data_ru, category_data_vn])   # combine for simpler summarization
    
    # create a summary statistics table
    category_summary = summary_stats(category_data)
    
    # store data without outliers in separate variables for plotting
    category_data_ru_no_out = category_data_ru.loc[~category_data_ru['is_outlier']]
    category_data_vn_no_out = category_data_vn.loc[~category_data_vn['is_outlier']]
    
    # add plots (use no_outliers data)
    # calculate number of bins as square root of number of data points
    bins_ru = 1 if len(category_data_ru_no_out) == 0 else m.ceil(m.sqrt(len(category_data_ru_no_out)))
    bins_vn = 1 if len(category_data_vn_no_out) == 0 else m.ceil(m.sqrt(len(category_data_vn_no_out)))

    # plot two histograms together
    fig, (ax1, ax2) = plt.subplots(2, 1)
    category_data_ru_no_out.normalized_price.hist(bins=bins_ru, ec='k', ax=ax1)
    ax1.set_title('Russia')
    category_data_vn_no_out.normalized_price.hist(bins=bins_vn, ec='k', ax=ax2)
    ax2.set_title('Vietnam')
    plt.tight_layout()

    # update the histogram in the dictionary
    data_summary[category]['hist'] = fig

    # trying to create a table description for histograms
    hist_table = pd.DataFrame(columns=['country', 'product_category', 'bin_min', 'bin_max', 'bin_count'])
    min_ru = category_data_ru_no_out.normalized_price.min()
    max_ru = category_data_ru_no_out.normalized_price.max()
    diff_ru = max_ru - min_ru
    bin_bound_ru = [min_ru] + [min_ru + (x+1) * diff_ru/bins_ru for x in range(bins_ru-1)] + [max_ru+1]
    for i in range(bins_ru):
        bin_count = category_data_ru_no_out.loc[(category_data_ru_no_out.normalized_price >= bin_bound_ru[i]) & (category_data_ru_no_out.normalized_price < bin_bound_ru[i+1])].normalized_price.count()
        hist_table.loc[len(hist_table)] = ['Russia', category, bin_bound_ru[i], bin_bound_ru[i+1], bin_count]
    min_vn = category_data_vn_no_out.normalized_price.min()
    max_vn = category_data_vn_no_out.normalized_price.max()
    diff_vn = max_vn - min_vn
    bin_bound_vn = [min_vn] + [min_vn + (x+1) * diff_vn/bins_vn for x in range(bins_vn-1)] + [max_vn+1]
    for i in range(bins_vn):
        bin_count = category_data_vn_no_out.loc[(category_data_vn_no_out.normalized_price >= bin_bound_vn[i]) & (category_data_vn_no_out.normalized_price < bin_bound_vn[i+1])].normalized_price.count()
        hist_table.loc[len(hist_table)] = ['Vietnam', category, bin_bound_vn[i], bin_bound_vn[i+1], bin_count]
    
    # bar and box plots
    category_barplot = category_summary.T.loc[['median']].plot.bar(rot=0).figure     # note: the summary table is transposed before applying barplot
    category_boxplot = category_data.boxplot(by='country', column='normalized_price', showfliers=False).figure
    plt.close('all')
    
    # update the dictionary entry
    data_summary[category].update({
        'data_ru': category_data_ru,
        'data_vn': category_data_vn,
        'summary': category_summary,
        'barplot': category_barplot,
        'boxplot': category_boxplot,
        'hist_table': hist_table
    })

    # data_summary[category]['data_ru'].to_csv(f'{category}_data_ru.csv', index=False)
    # data_summary[category]['data_vn'].to_csv(f'{category}_data_vn.csv', index=False)
    # data_summary[category]['summary'].to_csv(f'{category}_summary_stats.csv')
```
